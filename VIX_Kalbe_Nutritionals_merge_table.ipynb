```python
import pandas as pd
```

```python
customer_url = "https://raw.githubusercontent.com/asyhar22/kalbe_farma_ds/main/DATA/Case%20Study%20-%20Customer.csv"
store_url = "https://raw.githubusercontent.com/asyhar22/kalbe_farma_ds/main/DATA/Case%20Study%20-%20Store.csv"
product_url = "https://raw.githubusercontent.com/asyhar22/kalbe_farma_ds/main/DATA/Case%20Study%20-%20Product.csv"
transaction_url = "https://raw.githubusercontent.com/asyhar22/kalbe_farma_ds/main/DATA/Case%20Study%20-%20Transaction.csv"
```

```python
customer_table = pd.read_csv(customer_url, delimiter=";")
store_table = pd.read_csv(store_url, delimiter=";")
product_table = pd.read_csv(product_url, delimiter=";")
transaction_table = pd.read_csv(transaction_url, delimiter=";")
```

```python
customer_table.head(5)
```

```python
store_table.head(5)
```

```python
product_table.head(5)
```

```python
transaction_table.head(5)
```

```python
df = pd.merge(transaction_table, product_table, on="ProductID", how='left')
df = pd.merge(df, customer_table, on="CustomerID", how='left')
df = pd.merge(df, store_table, on="StoreID", how='left')
df.head(5)
```

```python
df.to_excel("merged_data.xlsx", index=False)
```
